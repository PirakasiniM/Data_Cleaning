```python
import pandas as pd
```

```python
df = pd.read_excel('/content/Customer Call List.xlsx')
df
```

 **Remove duplicates**

```python
df = df.drop_duplicates()
df
```

**Drop Columns**

```python
df = df.drop(columns = "Not_Useful_Column")
df
```


*   Remove '/' from lastname
*   Remove '.' from lastname
*   Remove '-' from lastname

If we use strip method as .strip() like this it will remove leading and trailing whitespace

```python
df['Last_Name'] = df['Last_Name'].str.strip('/')
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('.')
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('_')
df
```


*   remove any characters from the 'Phone_Number' column that are not alphanumeric.
*   format the phone numbers in the format xxx-xxx-xxxx.
*   converting each value in the 'Phone_Number' column to a string. It's redundant because the values are already expected to be strings.
*   format the phone numbers in the format xxx-xxx-xxxx.
*   remove instances of the string 'nan--' from the 'Phone_Number' column in a DataFrame.
*   if 'nan--' is the exact string representation of NaN in your DataFrame. If NaN is represented differently, you might need to adjust the code accordingly.

*In Pandas, [...] is used to create a character class or character set in regular expressions. Within the square brackets, you specify the characters you want to match.*

*^ inside a character class has a special meaning—it negates the character class. It means "match any character not in the specified set."*

*So, [^a-zA-Z0-9] in a regular expression means "match any character that is not an uppercase letter (A-Z), lowercase letter (a-z), or digit (0-9)."*

*In the context of the .str.replace() method in Pandas, when you use str.replace('[^a-zA-Z0-9]',''), it will replace any character in the 'Phone_Number' column that is not an alphanumeric character (letter or digit) with an empty string, effectively removing those characters from the strings.*

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

*splits the 'Address' column in a DataFrame (df) into three separate columns: 'Street_Address', 'State', and 'Zip_Code', based on the commas present in the address.*

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',2, expand=True)
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')

df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df
```

*fills all missing values in the DataFrame df with empty strings ('').*

```python
df = df.replace('N/a','')
#df = df.replace('NaN','')


df=df.fillna('')
df
```

***checks if the value in the "Do_Not_Contact" column is 'Y', and if so, drops the corresponding row from the DataFrame.***

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df

#Another way to drop null values
#df = df.dropna(subset="Phone_Number"), inplace=True)
```
